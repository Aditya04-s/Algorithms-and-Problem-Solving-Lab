# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
# Prob_4: Take Suitable example and write function that runs in logarithmic time. plot graph as well similar to above example.

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    operations = 0
    while left <= right:
        operations += 1
        mid = (left + right) // 2
        if arr[mid] == target:
            return operations
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return operations

n_values = list(range(1, 1001, 10))
time_values = []

for n in n_values:
    data = list(range(n))
    target_index = n // 2
    if n > 0:
        ops = binary_search(data, data[target_index])
    else:
        ops = 0
    time_values.append(ops)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Actual Operations')
plt.plot(n_values, [math.log2(n) if n > 0 else 0 for n in n_values], linestyle='--', color='red', label='O(log n) theoretical')
plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.legend()
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.

import math

def merge_sort_operations(arr):
    operations = 0
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        ops_left = merge_sort_operations(left_half)
        ops_right = merge_sort_operations(right_half)
        operations += ops_left + ops_right

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            operations += 1
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return operations

n_values = list(range(10, 1001, 10))
time_values = []

for n in n_values:
    import random
    data = [random.randint(0, n) for _ in range(n)]
    ops = merge_sort_operations(data.copy())
    time_values.append(ops)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Actual Operations')
plt.plot(n_values, [n * math.log2(n) for n in n_values], linestyle='--', color='red', label='O(n log n) theoretical')
plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.legend()
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
def polynomial_time_example(n):
    operations = 0
    for i in range(n):
        for j in range(n):
            operations += 1
    return operations

n_values = list(range(1, 101, 1))
time_values = []

for n in n_values:
    ops = polynomial_time_example(n)
    time_values.append(ops)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Actual Operations (O(n^2))')
plt.plot(n_values, [n**2 for n in n_values], linestyle='--', color='red', label='O(n^2) theoretical')
plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.legend()
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
operations_count = 0
def fibonacci_recursive_operations(n):
    global operations_count
    operations_count += 1
    if n <= 1:
        return n
    else:
        return fibonacci_recursive_operations(n-1) + fibonacci_recursive_operations(n-2)

n_values = list(range(1, 15))
time_values = []

for n in n_values:
    operations_count = 0
    fibonacci_recursive_operations(n)
    time_values.append(operations_count)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Actual Operations (Fibonacci)')
plt.plot(n_values, [2**n for n in n_values], linestyle='--', color='red', label='O(2^n) theoretical')
plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("O(2^n) Exponential Time Complexity (Recursive Fibonacci)")
plt.legend()
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
import math

operations_quick_sort = 0

def partition(arr, low, high):
    global operations_quick_sort
    pivot = arr[high]
    i = (low - 1)
    for j in range(low, high):
        operations_quick_sort += 1
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return (i + 1)

def quick_sort_operations(arr, low, high):
    global operations_quick_sort
    if low < high:
        pi = partition(arr, low, high)
        quick_sort_operations(arr, low, pi - 1)
        quick_sort_operations(arr, pi + 1, high)

n_values = list(range(10, 501, 10))
time_values_non_random_input = []
time_values_worst_case = []

for n in n_values:
    operations_quick_sort = 0
    data_non_random = list(range(n))
    if n > 1:
        data_non_random[0], data_non_random[n-1] = data_non_random[n-1], data_non_random[0]
    quick_sort_operations(data_non_random, 0, len(data_non_random) - 1)
    time_values_non_random_input.append(operations_quick_sort)

    operations_quick_sort = 0
    data_worst = list(range(n))
    quick_sort_operations(data_worst, 0, len(data_worst) - 1)
    time_values_worst_case.append(operations_quick_sort)

plt.figure(figsize=(12, 8))
plt.plot(n_values, time_values_non_random_input, label='Actual Operations (Quick Sort - Non-Random Input)')
plt.plot(n_values, [n * math.log2(n) for n in n_values], linestyle='--', color='green', label='O(n log n) theoretical')
plt.plot(n_values, time_values_worst_case, label='Actual Operations (Quick Sort - Worst Case)', color='purple')
plt.plot(n_values, [n**2 for n in n_values], linestyle=':', color='red', label='O(n^2) theoretical (Worst Case)')

plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("Quick Sort Time Complexity (without 'random' module)")
plt.legend()
plt.grid(True)
plt.show()
```
